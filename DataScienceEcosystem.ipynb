***Data Science Languages***     
Some examples ofcommonly used languages for Data Science are:
1. **Python**: General pourpose and open source programming language including powerful libraries for Data Science (NumPy, SciPy, Pandas, MatPlotLib,...). 
2. **R**: Statistical programming and free software language and powerful tool for data processing, manipulation and drawing.
3. **SQL**: Acronym for “Structured Query Language”, is a non-procedural language limited to querying and managing data.
4. **Other languages**: Java, JavaScript, Julia, C++, Scala are other usuful languages for Data Science.
 This is a non-comprehensive list and other languages can be used for data management.

***Data Science Libraries***    
Libraries (AKA frameworks) are a collection of functions and methods that allow you to perform many actions without writing the code. Some of the popular libraries used by Data Scientists include:   
1. **Numpy**: Python library for arrays and matrices management. 
2. **Pandas**: Python library providing tools for data structures manipulation and visualization. Pandas works with 2D tables called DataFrames.
3. **Matplotlib**: Powerful Python library for data visualization, including a full range of plots and graphs.
4. **Seaborn**: Python advanced library for data plots. Among others, includes heat maps, violin plots and time series.
5. **Sci-learn**: High-level Python library for Machine Learning models, including model definition, training, testing and validation. Features well known models for regression, classification or clustering.
6. **TensorFlow**: Low-level Python library for Deep Learning models production, deploymenty and operation.
7. **Keras**: High-level Python library for Deep Learning models, allowing Neural Networks definition, training and definition in a very simple-way.

It is remarkable that each programming language uses its own libraries, usually providing similar functionalities.
